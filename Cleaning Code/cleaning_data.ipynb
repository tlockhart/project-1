```python
import pandas as pd
```

```python
df_2017 = pd.read_csv('Resources/2017_merged_rushing_salary_finals.csv')
df_2017.head()
```

```python
df_2017.info()
```

```python
df_2018 = pd.read_csv('Resources/2018_merged_rushing_salary_finals.csv')
df_2018.head()
```

```python
df_2018.info()
```

```python
# concatenating the two dfs
runningback_df = pd.concat([df_2018, df_2017])
runningback_df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
runningback_df['Player'].value_counts()
```

```python
display(runningback_df[runningback_df['Player'] == 'Deonte Thompson'])
display(runningback_df[runningback_df['Player'] == 'Carlos Hyde'])
display(runningback_df[runningback_df['Player'] == 'Ty Montgomery'])
```

```python
runningback_df
```

```python
two_teamers = runningback_df.loc[runningback_df['Tm_x'] == '2TM']['Player']
duplicate_names = runningback_df[runningback_df['Player'].isin(two_teamers)]
duplicate_names_df = duplicate_names
```

```python
filtered_df = runningback_df[
  runningback_df['Tm_x'].str.contains("TM", na=False) |
  (~runningback_df['Player'].isin(two_teamers))
]
filtered_df['Player'].value_counts()
```

```python
display(filtered_df[filtered_df['Player'] == 'Michael Thomas'])
display(runningback_df[runningback_df['Player'] == 'Michael Thomas'])
display(runningback_df[runningback_df['Player'] == 'Chris Thompson'])
display(runningback_df[runningback_df['Player'] == 'Ryan Griffin'])
```

```python
# dropping oddball name duplicates
filtered_df = filtered_df.drop(index=[375, 465, 462])
```

```python
filtered_df = filtered_df.drop(columns=['Year_y', 'Tm_y'])
filtered_df = filtered_df.rename(columns={'Tm_x': 'Tm', 'Year_x': 'Year'})
filtered_df.head(10)
```

```python
filtered_df.columns
```

```python
filtered_df = filtered_df.rename(columns={
    'Rk': 'Rank', 'Tm': 'Team', 'G': 'Games Played', 'GS': 'Games Started',
    'Att': 'Attempts', 'Yds': 'Yards', 'TD': 'Touchdowns', '1D': ' First Downs',
    'Succ%': 'Success Percentage', 'Lng': 'Longest Run', 'Y/A': 'Yards per Attempt',
    'Y/G': 'Yards per Game', 'A/G': 'Attempts per Game', 'Fmb': 'Fumbles', 'Pos': 'Position',
    'cap_percent': 'Cap%'})
```

```python
filtered_df.head()
```

```python
# adding awards as individual columns
# PB: Proball
# AP: All Pro 
# OroY: Offensive rookie of the year
# OPoY: Offensive Player of the year
# CPoY: Comeback player of the year
filtered_df.loc[filtered_df['Awards'].notna()]
```
